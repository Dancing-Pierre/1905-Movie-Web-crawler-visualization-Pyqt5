```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('xuanyi.csv')
weight_range_counts = df.groupby('score').size()
plt.title('悬疑片在各评分区间柱状展示', fontproperties='SimHei', fontsize=14)
plt.bar(weight_range_counts.index, weight_range_counts.values)
plt.show()
```
